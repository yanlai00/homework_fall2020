```python
from tbparser.summary_reader import SummaryReader
import os
import matplotlib.pyplot as plt
```

```python
def read(logdir):
    reader = SummaryReader(logdir, tag_filter=['Eval_AverageReturn'])
    values = []
    for item in reader:
        values.append(item.value)
    return values
```

```python
root_dir = "/Users/yangyanlai/Desktop/homework_fall2020/hw2/data/"
subfolders = os.listdir(root_dir)
subfolders.remove(".DS_Store")
subfolders.sort()
```

```python
for subfolder in subfolders:
    print(subfolder)
```

```python
plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for subfolder in subfolders:
    if subfolder == "q2_b2000_r0.02_InvertedPendulum-v2_22-09-2020_22-13-25":
        logdir = os.path.join(root_dir, subfolder)
        values = read(logdir)
        batch_size = subfolder[4:8]
        lr = subfolder[10:14]
        label = f'batch_size={batch_size}, lr={lr}'
        plt.plot(values, label=label)

plt.title("The performance on Inverted Pendulum")
plt.xlabel("Number of Iterations")
plt.ylabel("Average return")
plt.legend()
plt.show()
```

```python
plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for subfolder in subfolders:
    if subfolder[:2] == "q2":
        logdir = os.path.join(root_dir, subfolder)
        values = read(logdir)
        batch_size = subfolder[4:8]
        lr = subfolder[10:14]
        label = f'batch_size={batch_size}, lr={lr}'
        plt.plot(values, label=label)

plt.title("The effect of batch size and learning rate")
plt.xlabel("Number of Iterations")
plt.ylabel("Average return")
plt.legend()
plt.show()
```

```python
plt.figure(num=None, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
for subfolder in subfolders:
    if subfolder[:2] == "q1":
        if subfolder.find("_sb_") != -1:
            logdir = os.path.join(root_dir, subfolder)
            values = read(logdir)
            label = ""

            no_rtg = subfolder.find("_no_rtg_")
            dsa = subfolder.find("_dsa_")
            
            label += "sb_"

            if no_rtg == -1:
                label += "rtg_"
            else:
                label += "no_rtg_"

            if dsa == -1:
                label += "na"
            else:
                label += "dsa"

            plt.plot(values, label=label)

plt.title("The effect of advantage standardization and reward to go with small batch size")
plt.xlabel("Number of Iterations")
plt.ylabel("Average return")
plt.legend()
plt.show()
```
